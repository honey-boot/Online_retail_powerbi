```python
df=df.drop_duplicates()
```

```python
df=df.dropna(subset=["Description"])
```

```python
df["Customer ID"]=df["Customer ID"].fillna("Unknown Product")
```

```python
df=df[~df["Invoice"].astype(str).str.startswith("C")]
```

```python
df["Quantity"].describe()
```

```python
df[df['Quantity'] <= 0]
```

```python
df=df[df["Quantity"]>0]
```

```python
df[df['Price'] <= 0]
```

```python
df=df[df["Price"]>0]
```

```python
df.shape
```

```python
#Feature engineering

df["Revenue"]=df["Quantity"] * df["Price"]
```

```python
df["YearMonth"] = df["InvoiceDate"].dt.to_period("M")
```

```python
df[["InvoiceDate", "YearMonth"]].head()
```

```python
#Calculate Monthly Revenue

monthly_sales = (
    df.groupby("YearMonth")["Revenue"]
      .sum()
      .reset_index()
)

monthly_sales.head()
```

```python
pip install  matplotlib
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(monthly_sales["YearMonth"].astype(str),
         monthly_sales["Revenue"])

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Which countries generate the highest revenue?

country_sales = (
    df.groupby("Country")["Revenue"]
      .sum()
      .sort_values(ascending=False).reset_index()
)

country_sales.head(10)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

top10 = country_sales.head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country"], top10["Revenue"])

plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Revenue")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Which Products Generate the Highest Revenue?

product_sales = (
    df.groupby("Description")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

product_sales.head(10)
```

```python
import matplotlib.pyplot as plt

top10 = product_sales.head(10)

plt.figure(figsize=(12,6))

plt.bar(top10["Description"], top10["Revenue"])

plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
#Who are the Top 10 Customers by Revenue?

customer_sales = (
    df.groupby("Customer ID")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

customer_sales.head(10)
```

```python
import matplotlib.pyplot as plt

top10 = customer_sales.head(10)

plt.figure(figsize=(10,6))

plt.bar(top10["Customer ID"].astype(str),
        top10["Revenue"])

plt.title("Top 10 Customers by Revenue")
plt.xlabel("Customer ID")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Which Product Categories Contribute the Most Revenue?

top_quantity = (
    df.groupby("Description")["Quantity"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

top_quantity.head(10)
```

```python
import matplotlib.pyplot as plt

top10 = top_quantity.head(10)

plt.figure(figsize=(12,6))

plt.bar(top10["Description"], top10["Quantity"])

plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
